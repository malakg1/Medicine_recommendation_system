```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('medical_data_edited.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['Name','DateOfBirth','Causes','Gender'], axis=1, inplace=True)
```

```python
symptoms_split = data['Symptoms'].str.split(',')

unique_symptoms = set()

for symptom_list in symptoms_split:
    unique_symptoms.update(symptom_list)

for symptom in unique_symptoms:
    data[symptom.strip()] = 0

for idx, symptom_list in enumerate(symptoms_split):
    for symptom in symptom_list:
        data.at[idx, symptom.strip()] = 1

data.drop('Symptoms', axis=1, inplace=True)
```

```python
data['Medicine'].fillna('', inplace=True)


medicine_split = data['Medicine'].str.split(',')


unique_medicines = set()

for medicine_list in medicine_split:
    unique_medicines.update(medicine_list)

for medicine in unique_medicines:
    data[medicine.strip()] = 0
    
target = unique_medicines

for idx, medicine_list in enumerate(medicine_split):
    for medicine in medicine_list:
        data.at[idx, medicine.strip()] = 1
        
data.drop('Medicine', axis=1, inplace=True)
```

```python
data['Disease'].fillna('', inplace=True)


Disease_split = data['Disease'].str.split(',')


unique_Disease = set()

for Disease_list in Disease_split:
    unique_Disease.update(Disease_list)

for Disease in unique_Disease:
    data[Disease.strip()] = 0
    
target_dis = unique_Disease

for idx, Disease_list in enumerate(Disease_split):
    for Disease in Disease_list:
        data.at[idx, Disease.strip()] = 1
        
data.drop('Disease', axis=1, inplace=True)
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
data.head()
```

```python
frequent_itemsets = apriori(data, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)
```

```python
rules
```

```python
# Input gathering
user_symptoms = input("Enter patient's information and symptoms separated by comma: ").split(',')

# Filter rules based on user symptoms
filtered_rules = rules.copy()
for index, row in filtered_rules.iterrows():
    antecedents = set(row['antecedents'])
    if not antecedents.intersection(user_symptoms):
        filtered_rules.drop(index, inplace=True)

# Extract recommended medicines and their confidence from filtered rules
recommended_medicines = set()  # Use a set to automatically remove duplicates
recommended_disease = set() 
if not filtered_rules.empty:
    for index, row in filtered_rules.iterrows():
        confidence = row['confidence']  # Assuming 'confidence' is a column in your DataFrame
        consequent_set = row['consequents']
        for med in consequent_set:
            if med != '' and med in target:
                recommended_medicines.add((med, confidence)) # Adding as tuple to preserve both medicine and confidence
            elif med != '' and med in target_dis:
                recommended_disease.add((med, confidence))
        


# Convert the set back to a list
recommended_medicines = list(recommended_medicines)
recommended_disease = list(recommended_disease)

# Print filtered recommended medicines along with confidence
print("\nRecommended Medicines:")
if recommended_medicines:
    # Sort recommended medicines by confidence before printing
    recommended_medicines.sort(key=lambda x: x[1], reverse=True)
    for index, (medicine, confidence) in enumerate(recommended_medicines):
        if index < 4:
            print(f"{medicine} (Confidence: {confidence})")
        else:
            break
else:
    print("No medicines recommended based on input symptoms.")

    
print("\nRecommended Disease:")
if recommended_disease:
    # Sort recommended disease  by confidence before printing
    recommended_disease.sort(key=lambda x: x[1], reverse=True)
    for index, (disease, confidence) in enumerate(recommended_disease):
        if index < 2 and confidence == 1.0:
            print(f"{disease} (Confidence: {confidence})")
        else:
            break
else:
    print("No disease recommended based on input symptoms.")
```
